```python
def colorMap(f):
    """Graphs the color maps of a function of 2 variables, f(x,y)"""
    x = np.linspace(0, 2, 19)
    y = np.linspace(0, 2, 19)
    x, y = np.meshgrid(x, y)
    fig,ax3 = plt.subplots(1)
    fig.suptitle('Actual function', fontsize = 20)
    plot = ax3.contourf(x,y,f(x,y))
    cbar = fig.colorbar(plot)
    return plot
    
```

```python
def colorMapList(f,a,b,c,d,n,points): #need to fix absolute color issue
    """USED FOR MESH ANALYSIS"""
    """Plots the colormap of a region [a,b] x [c,d] split into n^2 subareas, with a dataframe of one
    point chosen for each subarea, using the function f"""
    w = b - a
    h = d - c
    xlist = np.arange(2/(2*(n)),w,2/(n)) #center x coords
    ylist = np.arange(2/(2*(n)),h,2/(n)) #center y coords
    X,Y = np.meshgrid(xlist,ylist)
    pointList = points[['x','y']].values.tolist() #coords you are estimating the function with
    pointList = np.array(pointList)
    z = []
    for i in pointList:
        z.append(f(i[0],i[1])) #calculate values of the function from your pointList
    z = np.array(z)
    z = np.reshape(z,(n,n))
    fig,ax2 = plt.subplots(1)
    fig.suptitle('Experimental Result', fontsize=20)
    plot = ax2.contourf(Y,X,z) #plotting your estimations
    cbar = fig.colorbar(plot)
    return plot
    
```

```python
def compare(f,a,b,c,d,n,points):
    """Plots 2 figures - one of the color map of f, and one of the color map of a rectangle [a,b] x [c,d], split
    into n^2 subareas, using the list of points to estimate the color map"""
    #fig, axes = plt.subplots(nrows=2, ncols=2)
    q = plt.figure(1)
    colorMapList(f,a,b,c,d,n,points)
    #q.show()
    p = plt.figure(2)
    colorMap(f)
    plt.show()
```

```python
def colorize(f,points,a,b,c,d):
    """Given a function, values of the vertical Lines (x-mesh analysis) and a region
    [a,b] x [c,d] this function will make a contour plot of it"""
    length = points.shape[0] #number of rows in points
    xvals = np.array(points['Old X'], copy = True) #old x vals - just the center
    xvals = xvals.tolist()
    xvals.insert(0,a) #not sure why you need this
    xlist = xvals[:]
    xlist = np.array(xlist)
    yvals = np.array(points['Old Y'], copy = True)
    yvals = yvals.tolist()
    yvals.insert(0,c) #not sure why you need this either
    ylist = yvals[:]
    ylist = np.array(ylist)
    X,Y = np.meshgrid(xlist,ylist)
    pointList = points[['x','y']].values.tolist()
    pointList = np.array(pointList)
    z = []
    x = 0
    while (x < len(pointList)):
        for i in pointList:
            z.append(f(i[0],i[1])) #calculate the z values that you actually know - make array in right directions
        x+=1
    z = np.array(z)
    z = np.reshape(z,(length,length)) #reshape so right dimensions
    fig,ax2 = plt.subplots(1)
    ax2.pcolormesh(Y,X,z)
    
```
